```python
import os
import random
import numpy as np
import pandas as pd

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get all the "Hits" data files
root_dir = os.path.join('..', '..', '..', 'data', 'pdg13-n1-1GeV-0.5eta')
hit_files = sorted([file for file in os.listdir(root_dir) if file.endswith("-hits.csv")])
```

### Investigate file 5 files at random

```python
n = 5
colors = ['blue', 'cyan', 'yellow', 'magenta', 'red']
random_hit_files = sorted(random.sample(hit_files, n))
dfs = [pd.read_csv(os.path.join(root_dir, file)) for file in random_hit_files]
```

### Plots

- Plot in x/y plane (transverse view)

```python
plt.figure(figsize=(12, 7), dpi=80)
for i, file in enumerate(random_hit_files):
    plt.scatter(x=dfs[i]['tx'], y=dfs[i]['ty'], label=file, color=colors[i])
    
plt.legend(loc='upper left')
plt.savefig('transverse_view_plot.png')
```

- Plot in r/z plane (longitudinal view)

```python
plt.figure(figsize=(12, 6), dpi=80)
for i, file in enumerate(random_hit_files):
    r = np.sqrt(np.square(dfs[i]['tx']) + np.square(dfs[i]['ty']))
    plt.scatter(x=r, y=dfs[i]['tz'], label=file, color=colors[i])
    
plt.legend(loc='upper left')
plt.savefig('longitudinal_view_plot.png')
```

- 3D Plot

```python
fig = plt.figure(figsize=(12, 6), dpi=80)
ax = Axes3D(fig)

for i, file in enumerate(random_hit_files):
    ax.scatter(dfs[i]['tx'], dfs[i]['ty'], dfs[i]['tz'], label=file, color=colors[i])
    
ax.set_xlabel('tx')
ax.set_ylabel('ty')
ax.set_zlabel('tz')
ax.legend(loc='upper left')
plt.show()
```
